```python
import pandas as pd

data1 = pd.read_csv("general_data.csv")
data1
```

```python
data1['TotalWorkingYears']=data1['TotalWorkingYears'].fillna(11.28)
data1.columns
```

```python
data1.Attrition.eq('Yes').mul(1)
```

```python
from scipy.stats import pearsonr 
```

```python
stats,p = pearsonr(data1.YearsAtCompany,data1.Age)
print(stats,p)
if(p<0.05):
    print("Reject Null Hypothesis")
else:
    print("Accept Null Hypothesis")
```

```python
stats,p = pearsonr(data1.DistanceFromHome,data1.Age)
print(stats,p)
if(p<0.05):
    print("Reject Null Hypothesis")
else:
    print("Accept Null Hypothesis")
```

```python
stats,p = pearsonr(data1.Age,data1.TrainingTimesLastYear)
print(stats,p)
if(p<0.05):
    print("Reject Null Hypothesis")
else:
    print("Accept Null Hypothesis")
```

```python
stats,p = pearsonr(data1.Education,data1.JobLevel)
print(stats,p)
if(p<0.05):
    print("Reject Null Hypothesis")
else:
    print("Accept Null Hypothesis")
```
